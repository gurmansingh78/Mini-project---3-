# 1. Problem Statement:
* Find all the subsets from a set of numbers whose sum is zero.
* Constraint: Subset size must be 5
* Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
import itertools

# Given set
numbers = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

# Convert to list (optional, for consistent ordering)
number_list = list(numbers)

# Generate all 5-element combinations
five_element_subsets = itertools.combinations(number_list, 5)

# Filter subsets with sum = 0
zero_sum_subsets = [subset for subset in five_element_subsets if sum(subset) == 0]

# Print results
for subset in zero_sum_subsets:
    print(subset)

# Optional: print count
print(f"\nTotal subsets with sum 0: {len(zero_sum_subsets)}")
```

#2. Problem Statement:
* Find all the subsets from a set of numbers whose sum is zero.
* Constraint: Subset size must be 3 to 6 only
*Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}


```python
import itertools

# Given set
numbers = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

# Convert to list for combinations
number_list = list(numbers)

# Store all valid subsets
zero_sum_subsets = []

# Check combinations of size 3 to 6
for size in range(3, 7):
    for subset in itertools.combinations(number_list, size):
        if sum(subset) == 0:
            zero_sum_subsets.append(subset)

# Print the results
for subset in zero_sum_subsets:
    print(subset)

# Print total count
print(f"\nTotal subsets with sum 0 (size 3 to 6): {len(zero_sum_subsets)}")
```
